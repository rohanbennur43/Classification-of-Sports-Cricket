Function to plot the spectrogram

```python
def plot_spectrogram(Y, sr, hop_length, y_axis="linear"):
    plt.figure(figsize=(25, 10))
    librosa.display.specshow(Y,
                             sr=sr,
                             hop_length=hop_length,
                             x_axis="time",
                             y_axis=y_axis)
    plt.colorbar(format="%+2.f")
```

```python

FRAME_SIZE = 2048
HOP_SIZE = 512
for dirname, _, filenames in os.walk('/content/drive/MyDrive/Capstone dataset/audio_data'):

    for filename in filenames:

        wav_file_name = os.path.join(dirname, filename)

        audio,sr=librosa.load(wav_file_name)

        S_scale = librosa.stft(audio, n_fft=FRAME_SIZE, hop_length=HOP_SIZE)
        Y_scale = np.abs(S_scale) ** 2
        Y_log_scale = librosa.power_to_db(Y_scale)
        plot_spectrogram(Y_log_scale, sr, HOP_SIZE)
        #plot_spectrogram(Y_log_scale, sr, HOP_SIZE,y_axis="log")#since we tend to percieve sound as logarthmic function instear of linear
        break
```
